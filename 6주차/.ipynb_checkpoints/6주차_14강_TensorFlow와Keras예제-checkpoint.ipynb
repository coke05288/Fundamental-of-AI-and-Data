#### DataFrame 원하는 행, 열 접근

```python
# 조건식에 부합하는 데이터 가져오기
basket_ball[(basket_ball['3P']> 3.0) & (basket_ball['2P'] >3.0)]
```

#### DataFrame 열과 행 추가, 삭제

```python
# 열 추가
basket_ball['3P_double'] = basket_ball['3P'] * 2
basket_ball['3P_tripple'] = basket_ball['3P_double'] + basket_ball['3P']
basket_ball.head()
```

```python
# 행추가
basket_ball.loc['new_row'] = basket_ball.loc[198] + basket_ball.loc[199]
basket_ball.tail()
```

```python
# 원하는 위치에 행추가
basket_ball.insert(5, '3P12', basket_ball['3P']*12)
basket_ball.head()
```

```python
# 열 삭제
basket_ball.drop('3P12', axis = 1, inplace=True)
basket_ball.head()
```

```python
# 행 삭제
basket_ball.drop([0, 1], axis = 0, inplace= True)
basket_ball.head()
```

```python
basket_ball.corr()
```

## 주요 사용함수

- 깔끔하지 않은, 누락된 데이터들을 정제(클렌징)하기 위해 **데이터 전처리**를 해야한다.

NaN 값 확인 : `info()`, `isnull` or `isna()`\
NaN 삭제 : `dropna(), dropna(axis=1)`, `dropna(subset=['열이름',...]`\
NaN을 다른 데이터로 대체 : `fillna()`\

- 데이터를 분석에 용이하게 그룹화를 할 수 있다.

groupby('열이름') 함수\
groupping 함수 : `count()`, `sum()`, `mean(), std(), var()` 등

### 데이터 전처리

```python
import pandas as pd
import numpy as np

# data 출처 : https://www.kaggle.com/hssh97/titanicdeteset-traincsb/data
train_data = pd.read_csv('./train.csv')
train_data.head()
```

```python
train_data.shape
```

#### NaN 값 확인

- info 함수를 통하여 개수 확인

```python
train_data.info()
```

```python
train_data.isnull().head()
```

```python
train_data[['Age']].isnull().head()
```

#### NaN 처리 방법

- NaN 데이터 삭제 dropna 함수 - NaN이 있는 행 단위로 삭제

```python
train_data.dropna().tail()
```

```python
train_data
```

```python
train_data.dropna(subset=['Age']) # 'Age'열에 대해서만 NaN이 있는 행 삭제.
```

```python
train_data.dropna(axis=1) # NaN이 있는 열단위로 삭제
```

#### NanN 값 대체하기 : fillna 함수

```python
train_data.mean()
```

```python
train_data.fillna(train_data.mean()).head() # ()안에 넣어준 특정한 값으로 채움, 평균으로 대체
```

```python
train_data['Age'].mean()
```

```python
train_data[['Age']].fillna(train_data['Age'].mean())
```

```python
train_data['Age'] = train_data[['Age']].fillna(train_data['Age'].mean())
train_data.tail()
```

### 데이터 그룹화

#### groupby 함수

- 열의 이름으로 그룹화.
- groups 속성 : 각 그룹과 그룹에 속한 index를 dict 형태로 표현

```python
class_group = train_data.groupby('Pclass')
class_group
```

```python
class_group.groups
```

```python
gender_group = train_data.groupby('Sex')
gender_group.groups
```

#### groupping 함수

- 그룹 데이터에 적용 가능한 통계 함수 (NaN은 제외하여 연산(
- count : 데이터 개수
- sum - 데이터 합
- mean, std, var : 평균, 표준편차, 분산

```python
class_group.count()
```

```python
class_group.sum()
```

```python
class_group.mean()[['Survived']]
```

```python
gender_group.mean()[['Survived']]
```

#### 복수 columns로 groupping 하기

- groupby에 column 리스트를 전달
- 통계함수를 적용한 결과는 multiindex를 갖는 dataframe

##### 클래스와 성별에 따른 생존률 구해보기

```python
cg_group = train_data.groupby(['Pclass', 'Sex']).mean()
cg_group.index
```

```python
cg_group[['Survived']]
```

```python
cg_group.loc[[(2, 'female')]]
```
