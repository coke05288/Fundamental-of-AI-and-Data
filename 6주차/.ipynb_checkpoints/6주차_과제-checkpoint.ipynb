```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-3, 3, 0.1)
y1 = 10*x**3 - x
y2 = 10**3*np.cos(x)**2

plt.plot(x, y1, 'b', x, y2, 'r-.')
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph title')
plt.legend(['y1', 'y2'], loc = 'lower right')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier # KNN Model
from sklearn import svm
from sklearn import metrics
```

```python
df = pd.read_csv('./basketball_stat.csv')
data_df = df[0:200]
data_df
```

```python
train, test = train_test_split(data_df, test_size = 0.2)
print("train data", train.shape)
print("teset data", test.shape)
```

```python
train_data_df = train[['3P', '2P', 'BLK']]
train_label_df = train[['Pos']]
train_data = train_data_df.values # 값만 가져오기(2차원)
train_label = train_label_df.values.ravel() # 값 부분을 1차원으로
```

```python
train_data
```

```python
train_label
```

```python
test_data_df = test[['3P', '2P', 'BLK']]
test_label_df = test[['Pos']]
test_data = test_data_df.values # 값만 가져오기(2차원)
test_label = test_label_df.values.ravel() # 값 부분을 1차원으로
```

```python
test_data
```

```python
test_label
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(train_data, train_label)
```

```python
clf = svm.SVC(C = 1, gamma = 1)
clf.fit(train_data, train_label)
```

```python
pred_knn = knn.predict(test_data)
pred_knn
```

```python
pred_svm = clf.predict(test_data)
pred_svm
```
