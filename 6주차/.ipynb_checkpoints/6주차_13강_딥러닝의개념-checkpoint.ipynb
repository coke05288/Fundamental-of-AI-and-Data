# 딥러닝

- 머신러닝의 다양한 알고리즘의 한 종류인 **인공 신경망** 알고리즘을 이용한 기계학습
    - 인공신경망 : 인간의 뇌가 학습하는 방법을 수학적으로 모델링한 알고리즘
- 높은 정확도를 얻을 수 있다.
- 방대한 양의 학습 데이터와 높은 성능의 하드웨어가 필요함
- 학습 시간이 오래걸림

---

## 딥러닝 원리


### 선형 회귀 : 가장 훌륭한 예측선 긋기

<img src="./linear-regression.png">

$$y = ax + b\ $$
- 선형 회귀란, 임의의 직선을 그어 이에 대한 **평균 제곱근 오차(실제값 - 예측값)**를 구하고 이 값을 가장 작게 만들어주는 a와 b를 찾아가는 작업

$$ 평균 제곱근 오차 =  \sqrt{오차 제곱의 합 \over n}$$ 
 
    
### 다중 선형 회귀 : 가장 훌륭한 예측평면 찾기

<img src="./multiple-linear-regression.jpg">
$$y = a_1x_1 + a_2x_2  + b $$

### 로지스틱 회귀 : 합격과 불합격 판단하기

<img src="./logistic-regression.png">

- 로지스틱 회귀란, 학습 데이터에서 최적의 직선을 찾고 참(1)과 거짓(0)으로 분류하는 **S자 형태의 적절한 선**을 찾아가는 작업. -> 분류에 더 가깝다.
- y값을 다시 함수에 넣어 참, 거짓으로 분류.
- **딥러닝의 기본적인 요소**로 사용되는 알고리즘.

## 인공신경망(Artificial Neural Networks) 

### 인간의 뇌가 학습하는 과정

- 뉴런이 여러 수상돌기(input)를 통해 전기화학적인 신호를 받아들이고 축색돌기(output)에서 신호를 전송
- 뉴런 사이는 스냅스로 연결.
- 신호가 전달되기 위해서는 일정 기준(임계값 : 자극의총합)이상의 신호가 존재해야 함.


### 인공신경망 구조

- input(x)이 들어오면 가중치(w)가 곱해진다.
- 값들의 합(Y)이 임계값보다 크면 1을 다음 노드(뉴런)의 입력으로 보냄. (0은 출력이 없는 상태와 같다)
- 0과 1을 판단하는 함수를 활성화 함수(activation function)이라고 함.

<img src="./artificial-neural-networks.jpg">

#### 퍼셉트론
-  신경망을 이루는 가장 중요한 기본 단위
-  입력값과 활성화 함수를 사용해 출력 값을 다음으로 넘기는 가장 작은 신경망 단위

#### 활성화 함수
- 입력 신호를 받아 특정 값의 임계점을 넘어서는 경우에, 출력을 생성해주는 함수
- 임계값이 높으면 높을수록 분류 기준이 엄격함을 의미한다.
- sigmoid, ReLU, Leaky ReLU 등

#### 가중치(Weight)
- 입력신호가 결과 출력에 주는 영향도를 조절하는 매개변수

#### 편향(bias)
- 뉴런이 얼마나 쉽게 활성화 되느냐를 조정하는 변수.

## 딥러닝

### 다층 인공 신경망의 구조
- 노드가 서로 연결된 구조의 신경망
- 입력층에 학습시키고 시픙ㄴ 데이터를 입력
- 데이터들이 입력층, 중간층(은닉층), 출력층을 지나며 처리되고 최종결과를 출력.

<img src = "deep-learning.png">

### 딥러닝
- 딥러닝의 학습이란, 다층 신경망 구조에서 출력층에서의 오차가 최고사 되도록 각 노드를 연결하는 **접속 가중치와 편향 값을 학습**하는 과정.
- 노드가 서로 연결되어 있는 신경망 구조에서 **1개 이상의 은닉층을 두어(깊게할수록) 학습하면 정확도가 높은 결과**를 얻을 수 있다하여 사용 된 이름(딥)
- 은닉층을 거치면서 은닉층을 자동적으로 골라낸다.
- 머신러닝은 수동적으로 중요특징을 골라낸다.
