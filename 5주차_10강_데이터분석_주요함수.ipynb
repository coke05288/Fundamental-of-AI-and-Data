#### DataFrame 원하는 행, 열 접근

```python
# 조건식에 부합하는 데이터 가져오기
basket_ball[(basket_ball['3P']> 3.0) & (basket_ball['2P'] >3.0)]
```

```python
# 복수개의 열 가져오기
basket_ball[['Player', '3P', 'Pos']]
```

```python
# 행 이름으로 가지고 오고 싶을 때.
basket_ball.loc[[0]]
```

```python
# 행 위치로 가지고 오고 싶을 때.
basket_ball.iloc[[0, 3, 4]]
```

```python
# 행과 열 동시에 선택
basket_ball.loc[[3, 0, 7], ['Player', 'Pos', 'BLK']]
```

#### DataFrame 열과 행 추가, 삭제

```python
# 열 추가
basket_ball['3P_double'] = basket_ball['3P'] * 2
basket_ball['3P_tripple'] = basket_ball['3P_double'] + basket_ball['3P']
basket_ball.head()
```

```python
# 행추가
basket_ball.loc['new_row'] = basket_ball.loc[198] + basket_ball.loc[199]
basket_ball.tail()
```
