### Matplotlib 라이브러리

**MATLAB의 시각화 기능을 모델링해서 만들어진 파이썬 라이브러리**
- 선 그래프(plot), 산점도(scatter plot), 막대 그래프(bar chart), 히스토그램(histogram), 파이 그래프(pie chart) 등

### Plot 그래프

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
import matplotlib

matplotlib.rcParams['font.family'] = "Malgun Gothic"
matplotlib.rcParams['axes.unicode_minus'] = False
````

````python
data1 = [10, 14, 19, 20, 25]
# x 값은 생략 가능, 생략할 경우 0부터 y의 개수만큼 1씩 증가하는 값으로 할당.
plt.plot(data1)
plt.show()
````

#### x값과 y값이 모두 있는 2차원 선 그래프 그리기

````python
x = np.arange(-4.5, 5, 0.5)
y = 2*x**2

plt.plot(x,y)
plt.show()
````

#### 하나의 창에 여러 그래프 그리기

````python
x = np.arange(-5, 5, 0.1)
y1 = 2*x**2
y2 = 5*x + 30
y3 = 20*np.cos(x)**2

# plot.plot(x,y1,x,y2,x,y3)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)

plt.show()
````

#### 새로운 그래프 창에 그래프 그리기

````python
plt.figure(1)
plt.plot(x, y1)

plt.figure(2)
plt.plot(x, y2)

plt.figure(1)
plt.plot(x, y3)

plt.show()
````

#### 그래프 축의 범위 지정하기

````python
x = np.linspace(-4,4,100)
ya = x**3
yb = 10*x**2 -2

plt.plot(x, ya, x, yb)
plt.show()
````

````python
plt.plot(x,ya,x,yb)
plt.xlim(-1, 1)
plt.ylim(-3, 3)
plt.show()
````

#### 그래프 꾸미기
```py
fmt = '[color][line_style][marker]'
```

````python
plt.plot(x,y1,'y-',x,y2,'m-o',x,y3,'b--')
plt.xlim(-1,1)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph title')
plt.grid(True)
plt.legend(['data1','data2','data3'],loc = 'upper right')
plt.show()
````

### Scatter 그래프

- 두개의 요소로 이뤄진 데이터 집합의 관계를 시각화 하는데 사용.

```py
plt.scatter(x, y, [s = size_n, c = colors, marker = 'marker_string', alpha = alpha_f])
```

- 옵션을 지정하지 않을 경우 기본값 (s = 40, c = 'b', marker = 'o', alpha=1)
- alpha는 투명도(0에 가까울수록 투명함).

````python
height = [165, 170, 177, 159, 185]
weight = [62, 67, 79, 60, 94]

plt.scatter(height, weight, s = 80, c = ['red', 'black', 'yellow', 'blue', 'red'], marker = '*', alpha = 0.8)

plt.xlabel('Height(m)')
plt.ylabel('Weight(Kg)')
plt.title('Height & Weight')
plt.grid(True)

plt.show()
````

### Bar 그래프

- 여러 항목의 데이터를 서로 비교할 때 이용
```py
plt.bar(x, height, [width = width_f, color = colors, tick_label = tick_labels, align = 'center(default)'or 'edge'])
```

````python
member_ID = ['m_01', 'm_02', 'm_03', 'm_04']
before_ex = [30, 42, 40, 35]
after_ex = [27, 35, 40, 33]

n_data = len(member_ID)
index = np.arange(n_data)

colors = ['r', 'g', 'b', 'm']

plt.bar(index, before_ex, width = 0.4, color = colors, tick_label = member_ID)
plt.show()
````

````python
barWidth = 0.4

plt.bar(index, before_ex, width = barWidth, color = 'r',align = 'edge')
plt.bar(index + barWidth, after_ex, width = barWidth, color = 'm', align = 'edge', tick_label = member_ID)

plt.title('운동 전후 BMI 변화 비교')
plt.xlabel("회원 ID")
plt.ylabel("윗몸일으키기 횟수")
plt.legend(['운동 전', '운동 후'])

plt.show()
````
