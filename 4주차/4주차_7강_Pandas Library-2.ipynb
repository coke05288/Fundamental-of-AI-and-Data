### Import pandas, matolotlib 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
basket_ball = pd.read_csv('./basketball_stat.csv')
basket_ball.head()
```

### 새 Columns 추가하기

- [] 사용하여 추가하기.
- insert 함수 사용하여 원하는 위치에 추가하기.

#### [] 사용하여 추가

```python
basket_ball['3P_double'] = 0
basket_ball.head()
```

```python
basket_ball['3P_double'] = basket_ball['3P'] * 2
basket_ball[basket_ball['3P_double'] > 5]
```

```python
basket_ball['3P_tripple'] = basket_ball['3P_double'] + basket_ball['3P']
basket_ball[basket_ball['3P_tripple'] > 10]
```

#### insert 함수 사용하여 추가

```python
# insert(loc, column, value, allow_duplicates=False)
basket_ball.insert(3, '3P10', basket_ball['3P']*10)
basket_ball.head()
```

### 새 Row 추가하기

- .loc[] 사용하여 추가하기.

```python
basket_ball.loc['new_row'] = 0
basket_ball.tail()
```

```python
basket_ball.loc['new_row'] = basket_ball.loc[199] * 2
basket_ball.tail()
```

```python
basket_ball.loc['new_row'] = basket_ball.loc['new_row'] + basket_ball.loc[199]
basket_ball.tail()
```

### Columns 삭제하기

- drop 함수 사용하여 삭제, axis = 1.
- 원본을 삭제하기 위해서는 inplace = True 매개변수를 주어야 함.
- 리스트를 사용하여 멀티플 삭제 가능.

```python
basket_ball.drop('3P_tripple', axis =1)
```

```python
basket_ball.drop('3P_tripple', axis =1, inplace = True)
basket_ball.head()
```

```python
basket_ball.drop(['3P10', '3P_double'], axis = 1, inplace = True)
basket_ball.head()
```

### row 삭제하기

- drop 함수 사용하여 삭제, axis = 0
- 원본을 삭제하기 위해서는 inplace = True 매개변수를 주어야 함
- 리스트를 사용하여 멀티플 삭제 가능

```python
basket_ball.drop([0], axis = 0, inplace = True)
basket_ball.head()
```
