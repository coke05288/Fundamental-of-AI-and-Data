## 머신러닝

$$y = f(x) + \epsilon(오차항)\ $$  

- **대량의 데이터**와 알고리즘을 통해 컴퓨터에게 명시적으로 정의되지 않은 패턴을 학습
- 주어진 데이터 속에서 **데이터 특징**을 찾아내는 함수 f(모델)를 만드는 것.
- 학습으로 얻어진 정보를 기반으로 미래의 임의의 데이터에 대한 결과(값, 분포)를 예측.

**예시**
- 분류(Classification), 회귀(Regression), 군집(Clustering), 신경망(Neural Network) 등


#### 지도학습

- 학습할 데이터(X)와 명시적인 정답(Y : **Labeled Data**)을 이용해 데이터의 관계 특성과 분포를 학습하여 새로운 데이터에 대해 미래 결과를 예측하는 방법.

    - **분류(Classification)** : 학습 데이터(X)를 이용해 주어진 입력 값이 어떤 종류의 값(Y)인지 구별하는 것.
        - ex : 주가 등락, 비만 여부 등
    - **회귀(Regression)** : 학습 데이터(X)를 이용해 연속적인(숫자) 값(Y)을 예측하는 것.
        - ex : 몸무게, BMI 지수 등


#### 비지도학습

- 학습할 데이터(X)에 **정답(Y)이 없이** 입력 데이터만으로 숨겨진 특징이나 패턴을 학습을 통해 발견하는 방법.

    - **군집(Clustering)** : 사전 정보가 없는 주어진 데이터들에서 공통적인 요소들을 찾아 유사성이 높은 군들끼리 다른 그룹으로 나누는 방법.


#### 강화 학습

- 주어진 환경(**state**)에서 주체(**agent**)가 현재의 상태(state)를 관찰하며 선택할 수 있는 행동(**action**)들 중 가장 최대의 보상(**reward**)을 가져다 주는 행동이 무엇인지 학습하는 것.
- 지도학습과의 차이는 사람으로부터 학습을 받는 것이 아니라, 변화되는 환경으로부터 보상을 받아 학습한다는 차이를 가짐.
- 사람이 지식을 습득하는 방식 중 시행착오를 겪으며 학습하는 것과 매우 흡사하여 인공지능을 가장 잘 대표하는 모델.
- 프로세스 제어, 네트워크 관리, 로봇공학, 알파고, 자율 주행, 드론 등


#### 머신러닝 과정

1. 데이터 수집
2. **데이터 가공(분할)** - Training, Validation, Test 로 5:3:2의 비율로 분할.
3. 데이터 학습
    1. 모델 학습 -> **Training Data** 활용.
    2. 매개변수 조정
    3. 학습 방법(모델) 선택 -> **Validation Data** 활용
4. 최종성능지표 도출(**Test Data**활용) -> **모델 적합성 평가**
