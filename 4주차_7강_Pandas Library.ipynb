### Series Data

#### Series Data 생성하기

```python
import pandas as pd
import numpy as np
```

```python
# numpy의 arange함수로 배열 생성
a = np.arange(5)
# Series에 a 배열을 넣어 Series 객체 생성
s1 = pd.Series(a)

print(a)
print(s1)
```

```python
# 파이썬 list 자료형 생성
b = ['a', 'b', 'c', 'd', 'e']
# List로 Series 데이터 생성
s2 = pd.Series(b)
print(s2)
```

```python
# 인덱스 명시하기
s2 = pd.Series(a, b)
print(s2)
```

```python
c = [['a', 'b', 'c'], ['d', 'e', 'f']]
s3 = pd.Series(c)
print(s3)
```

```python
# 튜플로 생성하기
t = ('a', 'b', 'c', 'd', 'e')
s4 = pd.Series(t,b)
print(s4)
```

```python
# 딕셔너리로 Series 데이터 생성
d = {'국어': [100, 30], '영어':80, '수학':90}
s5 = pd.Series(d)
print(s5)
```

#### Series 데이터 속성과 인덱스 활용

```python
print(s5.index)
print(s5.values)
print(s5['국어'])
s5['영어'] = 90
print(s5['영어'])
```

### DataFrame 생성하기
- Series가 1차원이면 Datframe은 2차원으로 확대된 버전
- 2차원임으로 index는 row, column으로 구성됨
- 데이터분석과 머신러닝에서 데이터 변형을 위해 가장 많이 사용.

#### 데이터 파일 읽기
- NBA 농구선수들의 게임 기록 데이터 200개 항목
    - Player : 선수이름
    - Pos : 포지션
    - 3P : 한 경기 평균 3점 슛 성공 횟수
    - 2P : 한 경기 평균 2점 슛 성공 횟수
    - TRB : 한 경기 리바운드 성공 횟수
    - AST : 한 경기 어시스트 성공 횟수
    - STL : 한 경기 스틸 성공 횟수
    - BLK : 한 경기 블로킹 성공 횟수

#### 조건을 주어 원하는 Columns 선택하기

- & : and 둘다 만족
- | : or 둘 중 하나만 만족해도 ok

```python
basket_ball[basket_ball['3P'] > 1.0]
```

```python
basket_ball[(basket_ball['3P'] > 1.0) & (basket_ball['2P'] > 5)]
```

#### 원하는 행 선택하기

```python
basket_ball.loc[0]
```

```python
basket_ball.loc[[0]]
```

```python
basket_ball.iloc[0]
```

```python
basket_ball.iloc[[0]]
```

```python
basket_ball.iloc[[0, 4, 5]]
```
